#### 1. append
- 괄호 안의 값을 새로운 요소로 array 맨 끝에 객체로 추가
- 2개 이상의 시리즈 연결

```python
import pandas as pd
```

```python
df = pd.DataFrame({'성별':['여','남','여'],'나이':[21,24,28]},index=['A','B','C'])
```

```python
new_df = pd.DataFrame({'키':[163,175,159]},index=['A','B','C'])
```

```python
df.append(new_df)
```

```python
new_df = pd.DataFrame({'나이':[29,20,26]},index=['D','E','F'])
```

```python
df.append(new_df)
```

#### 3. drop_duplicates
- 중복되는 행을 제거하는 메서드
- 중복값이 없는 시리즈 반환

```python
mpg.head()
```

```python
mpg.fl.drop_duplicates() # 중복으로 사용되는 연료를 제거하고 유니크한 값만 반환한다
```

```python
mpg.drop_duplicates(['fl']) # 중복되는 값 중 처음 행만 표시
```

```python
mpg.drop_duplicates(['fl'],keep='last') # 중복되는 값 중 마지막 행만 표시
```

```python
mpg.drop_duplicates(['fl'],keep=False) 
# 중복되는 값은 모두 제거하고 유일한 값만 표시
```

#### 4. equals 
- 해당 값을 가진 요소가 있는지 확인해주는 메소드
- 두 객체에 동일한 요소가 포함되어 있는지 확인하는 메소드

```python
mpg.manufacturer.equals(mpg.model) 
# mpg의 manufacturer과 model 시리즈에는 동일한 요소가 포함되어있지 않다.
```

```python
mpg.cty.equals(mpg.cty)
```

##### 5. get_values
- 선택한 시리즈의 값 구하기 
- values 속성과 동일

- 버전에 따라 get_values 대신 to_numpy 이용!

```python
mpg[mpg['manufacturer']=='honda']['year'].to_numpy()
# manufacturer 명이 honda 안 year을 찾아준다.
```

```python
mpg[mpg['manufacturer']=='honda']['category'].to_numpy()
```

##### 6. isin
- 시리즈에 포함된 값이 있는지 확인하는 메서드
- 특정 문자로 찾을때는 isn 말고 str.contains() 이용!

```python
mpg[mpg.cty.isin([29])]
# cty가 29인 행 추출
```

##### 7. min, max

```python
mpg['hwy'].min()
```

#####  8. replace 
- 특정 값을 가진 시리즈를 교체

```python
mpg.replace(18,20)
# 18에 해당되는 모든 값이 20으로 바뀐다.
```

```python
mpg.replace({'hwy':26},20)
# hwy가 26에 해당되는 값만 20으로 바뀐다.
```

##### 9. sort_values
- 값을 오름차순으로 정렬한다.

```python
mpg.hwy.sort_values()
# 오름차순으로 정렬
```

```python
mpg.hwy.sort_values(ascending=False)
# 내림차순으로 정렬
```

```python
mpg.sort_values(by = 'hwy',ascending=False)
```
