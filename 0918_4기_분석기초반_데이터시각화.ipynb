### 데이터 시각화 (EDA) 
- 필수과정이며, 데이터 분석에 진행 중 초반에 시각화를 통해 새로운 인사이트를 찾는다.
- 시각화를 진행하기 위해서는 시각화에 목적이 명확해야 한다.
- 시각화의 목적, 이유가 명확하게 정해져야 시각화에 의미를 부여할 수 있다.
- 시각화는 간결해야 한다.

### 축 범위 설정

- xlim x축 설정
- ylim y축 설정
- left(해당 값 만큼 이동해서 시작) , top(최대값) 

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(0,2)
plt.ylim(3,5)
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(left = 1.5) #x축에만 작동
plt.ylim(top = 100) #y축에만 작동!
```

```python
plt.scatter([1,2,3],[4,5,6])
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(right = 1.5) 
plt.ylim(bottom = 100) 
```

### 눈금값 설정!

### 범례 설정!
- legend함수 사용한다!
- label 범례의 이름을 뜻함

```python
plt.scatter([1,2,3],[4,5,6], label ='x')
plt.scatter([-1,-2,-3],[4,5,6], label ='y')
plt.legend()#라벨의 범례 출력
```

- 범례의 위치를 바꿀 수 있는 인자
- left
- center
- right
- upper left
- upper center
- upper right
- lower left
- lower center
- lower right

```python
plt.scatter([1,2,3],[4,5,6], label ='x')
plt.scatter([-1,-2,-3],[4,5,6], label ='y')
plt.legend(loc ='center')#라벨의 범례 출력
```

### 그래프의 크기 및 스타일 설정하기!

- figure 그래프의 객체를 만드는 함수, figsize 순서대로 가로와 세로의 길이를 나타내는 튜플!

### 다양한 그래프 그리기!
- 선 그래프 그리기!

```python
plt.plot([1,2,3],[4,5,6], label ='line x')
plt.plot([1,2,3],[7,8,9], label ='line y')
plt.legend(loc='center')
```

### 색상 설정하기!

- color : black, gray yellow, brown, blue, rightblue
- 구글링해서 색상 더 찾기!! 

```python
plt.plot([1,2,3],[4,5,6],color ='red',label ='line x')
plt.plot([1,2,3],[7,8,9],color = 'blue', label ='line y')
plt.legend(loc='center')
```

### 선스타일, 두께 설정
- linestyle
- linewidth

```python
plt.plot([1,2,3],[4,5,6],linestyle='dashed',linewidth=5,color ='red',label ='line x')
plt.plot([1,2,3],[7,8,9],linestyle='dotted',linewidth=3,color = 'blue', label ='line y')
plt.legend(loc='center')
```

### 마커 설정도 가능하다!
- .(점), 'o', 's'사각형, '*'별, 'D'다이아몬드 등등~ 
- 구글링해서 찾아보기!!

- markersize, marker

```python
plt.plot([1,2,3],[4,5,6],marker = '*',markersize=10,color ='red',label ='line x')
plt.plot([1,2,3],[7,8,9],marker = 's',markersize=10,color = 'blue', label ='line y')
plt.legend(loc='center')
```

### 투명도 설정
- 데이터가 겹쳐져 있을 때 분포를 알아보기 쉽게 하려고 진행

```python
import numpy as np 
```

```python
x = np.random.normal(0,1,10000) #평균이 0, 분산이 1, 전체 데이터는 10000개
y = np.random.normal(0,1,10000)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, alpha = 0.2) #투명도를 보여주는 alpha
```
