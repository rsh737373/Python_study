### 시계열 데이터프레임 만들기
- datetine 판다스에서 제공하는 시계열 데이터 형식
- 시계열 데이터는 dtype이 꼭 datetime으로 변경해야 한다 
- why? datetime 문법이 따로 있고, 이 문법을 적용시키려면 dtype이 datetime으로 설정되어야 한다. 

```python
df = pd.read_csv(r'C:/Users/rsh73/Desktop/country_timeseries.csv')
```

```python
df
```

```python
df.info() # 문자열 데이터
```

```python
# object를 datetime으로 변경해서 시계열 형식으로 맞춤
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Date']
```

```python
df
```

```python
df[df['Date'].str.contains('2015')] 
# str.contains 는 Dtype이 datetime 일때는 오류!
```
