### 데이터 시각화 (EDA)
- 필수과정이며, 데이터 분석에 진행 중 초반에 시각화를 통해 새로운 인사이트를 찾는다.
- 시각화를 진행하기 위해서는 시각화에 목적이 명확해야 한다.
- 시각화의 목적, 이유가 명확하게 정해져야 시각화에 의미를 부여할 수 있다.
- 시각화는 간결해야 한다.

### 그래프의 기초와 환경 설정

```python
from matplotlib import pyplot as plt
```

### 그래프의 제목 및 축 이름 설정

```python
plt.scatter([1,2,3],[4,5,6])
plt.title('Hi') #그래프의 제목
plt.xlabel('x hi') #x축 제목
plt.ylabel('y hi') #y축 제목
```

### 축 범위 설정
- xlim : x축 설정
- ylim : y축 설정
- left(해당 값 만큼 이동해서 시작), top(최대값)

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(0,2) # x축 0~2 
plt.ylim(3,5) # y축 3~5
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(left=1) # x축으로 1 만큼 이동해서 시작 , x축에만 작동한다
plt.ylim(top = 5) # y축의 최댓값 , y축에만 작동한다
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(right = 2.25) # x축 끝에 점이 2.25
plt.ylim(bottom = 4.5) # y축의 최솟값
```

### 눈금값 설정! 
- xticks, yticks 사용

```python
plt.scatter([1,2,3],[4,5,6])
plt.xticks([1,2,3],['A','B','C']) # x축의 헤당밗을 특정 문자로 바꿀 수 있다!
plt.yticks([4,5,6],['a','b','c'])
```

### 범례 설정!
- legend 함수를 사용한다!
- label : 범례의 이름을 뜻한다!

```python
plt.scatter([1,2,3],[4,5,6],label='x')
plt.scatter([-1,-2,-3],[4,5,6],label='y')
plt.legend() # 라벨의 범례를 출력하는 함수
```

### 그래프의 크기 및 스타일 설정하기!
- figure 함수의 figsize
- figure 그래프의 객체를 만드는 함수, figsize 순서대로 가로와 세로의 길이를 나타내는 튜플!!

```python
plt.figure(figsize=(10,5)) # figsize=(x축의 길이, y축의 길이)
plt.scatter([1,2,3],[4,5,6],label='x')
```

### 다양한 그래프 그리기!
- 선 그래프

```python
plt.plot([1,2,3],[4,5,6],label='line x')
plt.plot([1,2,3],[7,8,9],label='line y')
plt.legend()
```

### 색상 설정하기!
- color : black, gray yellow, brown, blue, rightblue

```python
plt.plot([1,2,3],[4,5,6],color ='red',label ='line x')
plt.plot([1,2,3],[7,8,9],color = 'blue', label ='line y')
plt.legend()
```

### 선스타일, 두께 설정
- linestyle
- linewidth

```python
plt.plot([1,2,3],[4,5,6],linestyle='dashed',linewidth=5,color ='r',label ='line x')
plt.plot([1,2,3],[7,8,9],linestyle='dotted',linewidth=3,color = 'b', label ='line y')
plt.legend()
```

### 마커 설정도 가능하다!
- .(점), 'o'(동그라미), 's'(사각형), '*'(별), 'D'(다이아몬드) 
- markersize, marker

```python
plt.plot([1,2,3],[4,5,6],marker = '*',markersize=10,color ='r',label ='line x')
plt.plot([1,2,3],[7,8,9],marker = 's',markersize=10,color = 'b', label ='line y')
plt.legend()
```

### 산점도
- scatter
- scatter 에서는 markersize 대신 s 로 입력!!

```python
plt.scatter([1,2,3],[4,5,6],marker = '*',s=35, label ='x')
plt.scatter([-1,-2,-3],[4,5,6],marker = 's',s=15,label ='y')
```

### 투명도 설정
- 데이터가 겹쳐져 있을 때 분포를 알아보기 쉽게 하기 위해서

```python
import numpy as np 
```

```python
x = np.random.normal(0,1,10000) #평균이 0, 분산이 1, 전체 데이터는 10000개
y = np.random.normal(0,1,10000)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, alpha = 0.2) #투명도를 보여준다
```
